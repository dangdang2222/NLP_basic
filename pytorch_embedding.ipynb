###nn.Embedding()

단어 -> 정수 인덱스 -> 임베딩 층 -> 임베딩 벡터(밀집 벡터)

```python
train_data = 'you need to know how to code'

word_set = set(train_data.split())
word_set
```

```python
vocab = {token: i+2 for i, token in enumerate(word_set)}
vocab['<unk>']=0
vocab['<pad>']=1
```

num_embeddings : 임베딩 할 단어 개수 = 단어 집합 크기

embedding_dim : 임베딩 벡터 차원

padding_idx : 패딩을 위한 토큰 인덱스

```python
import torch.nn as nn
embedding_layer = nn.Embedding(num_embeddings=len(vocab), embedding_dim=3, padding_idx=1)
```

```python
print(embedding_layer.weight)
```

```python
from torchtext.legacy import data
```

```python
from torchtext.legacy import datasets
```

```python
TEXT = data.Field(sequential=True, batch_first=True, lower=True)
LABEL = data.Field(sequential=False, batch_first=True)
```

```python
trainset, testset = datasets.IMDB.splits(TEXT,LABEL)
```

```python
print(len(trainset),len(testset))
```

```python
vars(trainset[0])
```

torchtext에서 제공하는 pretrained된 워드 임베딩

glove.6B.300d를 사용하자

```python
from torchtext.vocab import GloVe
```

```python
TEXT.build_vocab(trainset, vectors=GloVe(name='6B',dim=300),max_size=10000,min_freq=10)
LABEL.build_vocab(testset)
```

```python
TEXT.vocab.stoi
```

```python
print(TEXT.vocab.vectors.shape)
```

임베딩 벡터는 총 10,002개가 존재하며 각 단어는 300차원


```python
print(TEXT.vocab.vectors[0])
print(TEXT.vocab.vectors[1])
print(TEXT.vocab.vectors[10])
```

```python
import torch.nn as nn
import torch
embedding_layer = nn.Embedding.from_pretrained(TEXT.vocab.vectors, freeze=False)
```
